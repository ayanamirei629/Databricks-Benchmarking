```python
%sql
SHOW CATALOGS;

USE CATALOG workspace;
USE SCHEMA default;

CREATE VOLUME IF NOT EXISTS tpch_export;
```

```python
tables = [
    "customer",
    "lineitem",
    "nation",
    "orders",
    "part",
    "partsupp",
    "region",
    "supplier",
]

base_path = "/Volumes/workspace/default/tpch_export"

for t in tables:
    df = spark.table(f"samples.tpch.{t}")
    (
        df.coalesce(1)  # single CSV file per table (OK for SF1)
          .write.mode("overwrite")
          .option("header", "true")
          .csv(f"{base_path}/{t}")
    )
    print(f"Exported {t} to {base_path}/{t}")
```
